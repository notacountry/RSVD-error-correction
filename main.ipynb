```python
import numpy as np

from test_matrices import (
    diagonal_known_spectrum,
    exact_low_rank,
    exponential_decay,
    polynomial_decay,
    run_test,
    signal_plus_noise,
)
```

```python
N     = 5000
K     = 555
P     = 20
SIGMA = np.array([10.0 / (i + 1) for i in range(K)])
SEED  = 42

# polynomial_decay
ALPHA = 1.0

# exponential_decay
BETA = 0.5

# signal_plus_noise
NOISE_LEVEL  = 1.0
```

```python
A, sigma_true = exact_low_rank(n=N, k=K, sigma=SIGMA, seed=SEED)
run_test("Exact low-rank", A, sigma_true, k=K, p=P, seed=SEED)
```

```python
A, sigma_true = diagonal_known_spectrum(n=N, k=K, sigma=SIGMA, seed=SEED)
run_test("Diagonal known spectrum", A, sigma_true, k=K, p=P, seed=SEED)
```

```python
A, sigma_true = polynomial_decay(n=N, k=K, alpha=ALPHA, seed=SEED)
run_test("Polynomial decay", A, sigma_true, k=K, p=P, seed=SEED)
```

```python
A, sigma_true = exponential_decay(n=N, k=K, beta=BETA, seed=SEED)
run_test("Exponential decay", A, sigma_true, k=K, p=P, seed=SEED)
```

```python
A, sigma_true = signal_plus_noise(n=N, k=K, sigma_signal=SIGMA, noise_level=NOISE_LEVEL, seed=SEED)
run_test("Signal plus noise", A, sigma_true, k=K, p=P, seed=SEED)
```
